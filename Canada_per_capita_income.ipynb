```python
# read csv file using pandas
df = pd.read_csv('canada_per_capita_income.csv')
# show first 5 rows
df.head()
```

```python
X, y = df[['year']], df['per capita income (US$)']
```

```python
# plot scatter graph for years and per capita income
plt.xlabel('years')
plt.ylabel('per capita income')
plt.scatter(X, y, color="red", marker="*")
plt.show()
```

```python
# linear regression fittting the values
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
# prediction function (the function takes only 2d array)
reg.predict(X)
```

```python
reg.predict([[2020], [2021], [2022], [2023], [2024], [2025], [2026]])
```

```python
# slope(or gradient) (weight)
reg.coef_
```

```python
# intercept (bias)
reg.intercept_
```

```python
# plot regression graph
plt.xlabel('years')
plt.ylabel('per capita income')
plt.scatter(X, y, color="red", marker="*")
plt.plot(X, reg.predict(X), color="blue")
plt.show()
```

```python
# Mean Square Error function
def MSE(y_true, y_predicted):
  return np.mean((y_true - y_predicted)**2)
```

```python
# value of MSE
mse = MSE(y, reg.predict(X))
mse
```

```python
# function for each cost (error)
def cost(y_true, y_predicted):
  cost_list = []
  each_cost = []
  for i in range(len(y_true)):
    cost = y_true[i] - y_predicted[i]
    cost_list.append(cost)
    each_cost.append(i)

  return cost_list, each_cost
```

```python
# value of cost
cost, each = cost(y, reg.predict(X))
```

```python
plt.plot(cost, each)
plt.show()
```
