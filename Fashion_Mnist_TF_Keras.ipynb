```python
# import needed libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
# get the dataset from keras
fashion_mnist = keras.datasets.fashion_mnist
```

```python
# load the dataset and divide into train and test
(X_train_full, Y_train_full), (X_test, Y_test) = fashion_mnist.load_data()
```

```python
# class names
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
```

```python
# min-max normalization
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0
```

```python
# samples for validation and train
X_valid = X_train_full[:6000]
X_train = X_train_full[6000:]

Y_valid = Y_train_full[:6000]
Y_train = Y_train_full[6000:]
```

```python
# for plot learning curves
def plot_learning_curve(history, title):
    plt.figure()
    plt.plot(history.history['loss'], label='Training loss')
    plt.plot(history.history['val_loss'], label='Validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# MLP neural network with one hidden layer with 100 neurons
model_for_1 = keras.models.Sequential([

    # converts a 2D image into a 1D vector
    keras.layers.Flatten(input_shape=[28, 28]),

    # Hidden layer with ReLU
    keras.layers.Dense(100, activation=tf.nn.relu),

    # Output layer with 10 neurons
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# settings how the model should learn
model_for_1.compile(
    # Loss function measures how far the model’s predictions are from the true labels
    loss='sparse_categorical_crossentropy',
    # The optimizer controls how the network updates its weights
    optimizer='adam',
    # Metrics are used to monitor the model’s performance.
    metrics=['accuracy']
)
```

```python
history_1 = model_for_1.fit(
    X_train, Y_train,
    epochs=10,
    validation_data=(X_valid, Y_valid)
)
```

```python
plot_learning_curve(history_1, "MLP with 1 Hidden Layer (100 neurons)")
```

Train a DNN on MNIST data set (hand written number from 0 to 9)

• The network consists of five hidden layers

• Each hidden layer has 100 neurons

• Use ELU activation function.

• Use Adam optimization

• Use early stopping

• Train your model ONLY with digits 0 to 4. NOTE: input is ONLY digits 0 to 4.

```python
# only didgits 0 to 4
train_mask = Y_train_full <= 4
test_mask = Y_test <= 4
```

```python
X_train = X_train_full[train_mask]
Y_train = Y_train_full[train_mask]
```

```python
X_test = X_test[test_mask]
Y_test = Y_test[test_mask]
```

```python
# min-max normalization
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# samples for validation and train
validation_size = int(0.1 * len(X_train))
X_valid, X_train = X_train[:validation_size], X_train[validation_size:]
Y_valid, Y_train = Y_train[:validation_size], Y_train[validation_size:]
```

```python
model_dnn = keras.models.Sequential([

    # converts a 2D image into a 1D vector
    keras.layers.Flatten(input_shape=[28, 28]),

    # Five Hidden layers with ELU activation
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),

    # Output layer 5 neurons with softmax
    keras.layers.Dense(5, activation=tf.nn.softmax)
])
```

```python
# settings how the model should learn
model_dnn.compile(
   # Loss function measures how far the model’s predictions are from the true labels
    loss='sparse_categorical_crossentropy',
    # The optimizer controls how the network updates its weights
    optimizer='adam',
    # Metrics are used to monitor the model’s performance.
    metrics=['accuracy']
)
```

```python
# early stop
early_stop = keras.callbacks.EarlyStopping(
    monitor='val_loss',    # Monitor validation loss
    patience=5,            # Stop if no improvement after 5 epochs
    restore_best_weights=True
)
```

```python
# Checkpoint callbacks
checkpoint_epoch = ModelCheckpoint(
    # filename includes epoch number
    'model_epoch_{epoch:02d}.h5',
    save_best_only=False,
    save_weights_only=False
)
```

```python
# the best model
checkpoint_best = ModelCheckpoint(
    'best_model.keras',
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)
```

```python
# train the model
history_dnn = model_dnn.fit(
    X_train, Y_train,
    epochs=10,
    validation_data=(X_valid, Y_valid),
    callbacks=[early_stop, checkpoint_epoch, checkpoint_best]
)
```

```python
# Evaluate on the test set
test_loss, test_acc = model_dnn.evaluate(X_test, Y_test)
print("Test accuracy:", test_acc)
```

```python
plot_learning_curve(history_dnn, "DNN with 5 Hidden Layers (100 neurons each)")
```

The model is overfitting

```python
# save the final model
model_dnn.save('final_model.keras')
```

```python
# only didgits 5 to 9
filter_from_5_to_9 = (Y_train_full >= 5) & (Y_train_full <= 9)
X_5_9_all = X_train_full[filter_from_5_to_9]
Y_5_9_all = Y_train_full[filter_from_5_to_9]
```

```python
# min-max normalization
X_5_9_all = X_5_9_all / 255.0
```

```python
# Select 100 images per digit
X_selected = []
Y_selected = []
```

```python
for digit in range(5, 10):
  # take first 100 images
    idx = np.where(Y_5_9_all == digit)[0][:100]
    X_selected.append(X_5_9_all[idx])
    Y_selected.append(Y_5_9_all[idx])
```

```python
# Concatenate all digits
X_selected = np.concatenate(X_selected)
Y_selected = np.concatenate(Y_selected)
```

```python
# Shuffle the dataset
perm = np.random.permutation(len(X_selected))
X_selected = X_selected[perm]
Y_selected = Y_selected[perm]
```

```python
# Split 80% training, 20% testing
split_idx = int(0.8 * len(X_selected))
X_train_new, X_test_new = X_selected[:split_idx], X_selected[split_idx:]
Y_train_new, Y_test_new = Y_selected[:split_idx], Y_selected[split_idx:]
```

```python
# Same DNN model architecture
model_of_c = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(100, activation=tf.nn.elu),
    keras.layers.Dense(5, activation=tf.nn.softmax)
])
```

```python
# settings how the model should learn
model_of_c.compile(
   # Loss function measures how far the model’s predictions are from the true labels
    loss='sparse_categorical_crossentropy',
    # The optimizer controls how the network updates its weights
    optimizer='adam',
    # Metrics are used to monitor the model’s performance.
    metrics=['accuracy']
)
```

```python
# reshape
X_train_new = X_train_new.reshape(-1, 28, 28)
X_test_new = X_test_new.reshape(-1, 28, 28)

# convert types
X_train_new = X_train_new.astype('float32')
X_test_new = X_test_new.astype('float32')
Y_train_new = Y_train_new.astype('int32') - 5  # shift labels to 0-4
Y_test_new = Y_test_new.astype('int32') - 5
```

```python
# train the model
history_of_c = model_of_c.fit(
    X_train_new, Y_train_new,
    epochs=20,
    validation_split=0.2  # 20% of training set for validation
)
```

```python
test_loss, test_acc = model_of_c.evaluate(X_test_new, Y_test_new)
print("Test accuracy on digits 5–9 subset:", test_acc)
```
